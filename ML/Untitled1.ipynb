```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

print(os.listdir())

df = pd.read_excel('ТЗ.xlsx')

# Просмотр структуры
df.columns
```

```python
# Преобразуем названия колонок к стандартному формату
df.columns = [col.strip() for col in df.columns]

# Переименуем нужную колонку
df.rename(columns={'Ставка в час': 'price'}, inplace=True)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=30, kde=False, color='steelblue')
plt.title('Распределение количества услуг по стоимости')
plt.xlabel('Стоимость за чел/час')
plt.ylabel('Количество услуг')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], bins=30, kde=True, color='teal')
plt.title('Гистограмма с линией нормального распределения')
plt.xlabel('Стоимость за чел/час')
plt.ylabel('Плотность и количество')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['price'], color='skyblue')
plt.title('Boxplot ставок за час')
plt.xlabel('Ставка, руб./ч')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=40, log_scale=True, color='darkorange')
plt.title('Распределение ставок (логарифмическая шкала)')
plt.xlabel('Ставка, руб./ч (лог)')
plt.ylabel('Количество')
plt.grid(True)
plt.show()
```
